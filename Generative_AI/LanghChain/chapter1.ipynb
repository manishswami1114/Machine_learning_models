### Getting Specific formats out of LLMs 

```python
from langchain_core.pydantic_v1 import BaseModel 

class AnswerWithJustification(BaseModel):
    '''An answer to the user's question along with just answer'''
    answer:str
    '''The answer to the user's question'''

    justification:str
    '''Justification for the answer'''


llm = OllamaLLM(model="deepseek-r1:1.5b",temperature=0)
structured_llm = llm.with_structured_output(AnswerWithJustification)
structured_llm.invoke("What weighs more, a pound of bricks or a pound of feathers")
```

### Imperative Composition 

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import chain

template = ChatPromptTemplate.from_messages([
    ('system','you are a helpful assistant.'),
    ('human','{question}'),
])

@chain
def chatbot(values):
    prompt = template.invoke(values)
    return model.invoke(prompt)

chatbot.invoke({"question":"which model providers offers LLMs"})
```

Streaming Support or Async support 

```python
@chain 
def chatbot(values):
    prompt = template.invoke(values)

    for token in model.stream(prompt):
        yield token
    
for part in chatbot.stream({
    "question":"which model provides offer LLMs"}):
    print(part)
```

```python
# for async 
@chain
async def chatbot(values):
    prompt = await template.ainvoke(values)
    return await model.ainvoke(prompt)

await chatbot.ainvoke({"question":"which model provides offers LLMs"})
```

### Declarative Composition 

```python
template = ChatPromptTemplate.from_messages([
    ('system','you are a helpful assistant'),
    ('human','{question}'),
])

chatbot = template | model

chatbot.invoke({"question":"which model providers offers LLMs"})
```

```python
# with invoke/stream/batch

chatbot = template|model
for part in chatbot.stream({
    "question":"which model providers offer LLMs?"}):
    print(part)
```

```python
# async 
chatbot = template|model 
await chatbot.ainvoke({
    "question","which model providers offer LLMs?"
})
```

All langchain componets share the same interface with `invoke`,`stream`,and `batch` method to handle various inputs and outputs. They can either be combined and executed imperatively by calling them directly or declaratively using LCEL
